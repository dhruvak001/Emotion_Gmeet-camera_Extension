```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
import os
import time
import copy
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from tqdm import tqdm
import os
# os.environ["CUDA_LAUNCH_BLOCKING"] = "1"
# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")


# Dataset paths
data_dir = 'affectnet/AffectNet'
train_dir = os.path.join(data_dir, 'train')
val_dir = os.path.join(data_dir, 'val')
test_dir = os.path.join(data_dir, 'test')

# Constants based on the modified implementation
BATCH_SIZE = 64
NUM_CLASSES = 8  # There are 8 emotion folders: 0,1,2,3,4,5,6,7
EPOCHS = 60
IMG_SIZE = 48  # Using 48x48 input as per FER2013 setup

# Define emotion mapping (if needed for interpretation)
emotion_classes = [0, 1, 2, 3, 4, 5, 6, 7]  # This maps folder names to emotion classes

# Data transformations: convert images to grayscale and resize to 48x48
data_transforms = {
    'train': transforms.Compose([
        transforms.Grayscale(num_output_channels=1),
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(20),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
        transforms.ToTensor(),
        transforms.Normalize([0.5], [0.5])
    ]),
    'val': transforms.Compose([
        transforms.Grayscale(num_output_channels=1),
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize([0.5], [0.5])
    ]),
    'test': transforms.Compose([
        transforms.Grayscale(num_output_channels=1),
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize([0.5], [0.5])
    ])
}

# Create datasets
image_datasets = {
    'train': datasets.ImageFolder(train_dir, data_transforms['train']),
    'val': datasets.ImageFolder(val_dir, data_transforms['val']),
    'test': datasets.ImageFolder(test_dir, data_transforms['test'])
}

# Function to plot class distribution
def plot_class_distribution(dataset):
    class_counts = [len([x for x in dataset.imgs if x[1] == i]) for i in range(len(dataset.classes))]
    plt.figure(figsize=(12, 6))
    plt.bar(dataset.classes, class_counts)
    plt.title('Class Distribution in Dataset')
    plt.xlabel('Emotion Classes')
    plt.ylabel('Number of Samples')
    plt.xticks(rotation=45)
    plt.show()

# Visualize class distribution for training set
plot_class_distribution(image_datasets['train'])

# Create dataloaders
dataloaders = {
    'train': DataLoader(image_datasets['train'], batch_size=BATCH_SIZE, shuffle=True, num_workers=4),
    'val': DataLoader(image_datasets['val'], batch_size=BATCH_SIZE, shuffle=False, num_workers=4),
    'test': DataLoader(image_datasets['test'], batch_size=BATCH_SIZE, shuffle=False, num_workers=4)
}

# Function to visualize augmented samples
def plot_augmented_samples(dataloader, n_samples=8):
    classes = dataloader.dataset.classes
    inputs, labels = next(iter(dataloader))
    inputs = inputs[:n_samples]
    labels = labels[:n_samples]
    
    # Unnormalize images
    mean = torch.tensor([0.5]).view(1, 1, 1, 1)
    std = torch.tensor([0.5]).view(1, 1, 1, 1)
    unnorm_images = inputs * std + mean
    
    plt.figure(figsize=(15, 6))
    for i in range(n_samples):
        plt.subplot(2, 4, i+1)
        img = unnorm_images[i].squeeze(0).numpy()  # remove channel dimension for grayscale
        plt.imshow(img, cmap='gray')
        plt.title(f"Label: {classes[labels[i]]}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

print("Training Samples with Augmentations:")
plot_augmented_samples(dataloaders['train'])

# ---------------------------
# Model setup: ResNet-50 based method from the paper
# ---------------------------
# Load the pretrained ResNet-50 model (pretrained on ImageNet)
model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)

# Modify the first convolutional layer to accept 1-channel (grayscale) input.
# This is done by averaging the pretrained weights across the RGB channels.
pretrained_weight = model.conv1.weight.clone()  # shape: [64, 3, 7, 7]
model.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
model.conv1.weight.data = pretrained_weight.mean(dim=1, keepdim=True)

# Replace the final fully connected layer to match NUM_CLASSES.
num_ftrs = model.fc.in_features
model.fc = nn.Sequential(
    nn.Linear(num_ftrs, 512),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(512, NUM_CLASSES)
)

model = model.to(device)

# ---------------------------
# Training visualization setup
# ---------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
plt.ion()  # Interactive mode on for live updates

# Enhanced training loop with visualizations
def train_model(model, criterion, optimizer, num_epochs=60):
    best_acc = 0.0
    best_model_wts = copy.deepcopy(model.state_dict())
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    
    for epoch in range(num_epochs):
        print(f'\nEpoch {epoch+1}/{num_epochs}')
        print('-' * 10)
        
        # Single progress bar per epoch covering both train and val phases
        epoch_bar = tqdm(total=len(dataloaders['train']) + len(dataloaders['val']),
                         desc=f'Epoch {epoch+1}',
                         position=0,
                         leave=True,
                         bar_format='{l_bar}{bar:20}{r_bar}{bar:-20b}')
        
        # Training phase
        model.train()
        running_loss = 0.0
        running_corrects = 0
        
        for inputs, labels in dataloaders['train']:
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            optimizer.zero_grad()
            
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            _, preds = torch.max(outputs, 1)
            
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)
            
            epoch_bar.set_postfix({'Phase': 'Train', 'Loss': loss.item()})
            epoch_bar.update(1)
        
        train_loss = running_loss / len(image_datasets['train'])
        train_acc = (running_corrects.double() / len(image_datasets['train'])).cpu().item()
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        
        # Validation phase
        model.eval()
        val_running_loss = 0.0
        val_running_corrects = 0
        
        with torch.no_grad():
            for inputs, labels in dataloaders['val']:
                inputs = inputs.to(device)
                labels = labels.to(device)
                
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                _, preds = torch.max(outputs, 1)
                
                val_running_loss += loss.item() * inputs.size(0)
                val_running_corrects += torch.sum(preds == labels.data)
                
                epoch_bar.set_postfix({'Phase': 'Val', 'Loss': loss.item()})
                epoch_bar.update(1)
        
        val_loss = val_running_loss / len(image_datasets['val'])
        val_acc = (val_running_corrects.double() / len(image_datasets['val'])).cpu().item()
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        epoch_bar.close()
        
        # Save best model weights based on validation accuracy
        if val_acc > best_acc:
            best_acc = val_acc
            best_model_wts = copy.deepcopy(model.state_dict())
            torch.save(model.state_dict(), 'best_model_weights.pth')
        
        print(f"Epoch {epoch+1} Train Loss: {train_loss:.4f} Acc: {train_acc:.4f} | Val Loss: {val_loss:.4f} Acc: {val_acc:.4f}")
    
    plot_training_curves(history)
    
    model.load_state_dict(best_model_wts)
    return model, history

def plot_training_curves(history):
    plt.figure(figsize=(15, 6))
    
    # Loss plot
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    # Accuracy plot
    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Train Accuracy')
    plt.plot(history['val_acc'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# ---------------------------
# Training execution
# ---------------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

model, history = train_model(model, criterion, optimizer, num_epochs=EPOCHS)

# ---------------------------
# Final evaluation visualization functions
# ---------------------------
def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

# Evaluation on test set
model.eval()
all_probs = []
all_labels = []
with torch.no_grad():
    for inputs, labels in dataloaders['test']:
        inputs = inputs.to(device)
        outputs = model(inputs)
        probs = torch.nn.functional.softmax(outputs, dim=1)
        all_probs.append(probs.cpu())
        all_labels.append(labels.cpu())

y_true = torch.cat(all_labels).numpy()
y_score = torch.cat(all_probs).numpy()
y_pred = np.argmax(y_score, axis=1)

# Classification report using folder names (0-7) as class labels
print(classification_report(y_true, y_pred, target_names=image_datasets['test'].classes))

# Confusion matrix
plot_confusion_matrix(y_true, y_pred, image_datasets['test'].classes)

# Save the final model
torch.save(model.state_dict(), 'emotion_model_weights.pth')
torch.save(model, 'emotion_model_full.pth')
```

```python
# ---------------------------
# Display test images with detected emotion labels
# ---------------------------
import matplotlib.pyplot as plt

# Load the saved full model
loaded_model = torch.load('emotion_model_full.pth')
loaded_model = loaded_model.to(device)
loaded_model.eval()

# Get one batch from the test dataloader
dataiter = iter(dataloaders['test'])
inputs, labels = next(dataiter)

# Move inputs to device and get predictions
inputs_device = inputs.to(device)
with torch.no_grad():
    outputs = loaded_model(inputs_device)
    probs = torch.nn.functional.softmax(outputs, dim=1)
    _, preds = torch.max(probs, 1)

# Unnormalize the images (for grayscale images normalized with mean=0.5 and std=0.5)
unnormalized_images = inputs * 0.5 + 0.5  # reverting normalization

# Plot images with predicted labels
plt.figure(figsize=(15, 6))
# Use up to 8 samples (or the batch size if smaller)
n_samples = min(8, inputs.size(0))
for i in range(n_samples):
    plt.subplot(2, 4, i+1)
    # Remove channel dimension and convert to numpy array
    img = unnormalized_images[i].squeeze(0).cpu().numpy()
    plt.imshow(img, cmap='gray')
    # Use the folder names as labels (i.e., '0','1',..., '7') from the dataset
    predicted_label = image_datasets['test'].classes[preds[i].cpu().item()]
    true_label = image_datasets['test'].classes[labels[i]]
    plt.title(f"Pred: {predicted_label}\nTrue: {true_label}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
